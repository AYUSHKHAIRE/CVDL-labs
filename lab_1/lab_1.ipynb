## Modify the position of a dog in an image by manipulating its coordinates using NumPy.

```python
if img is None:
    raise ValueError("Image not found.")
dog_bbox = (693, 142, 521, 848) 
dog_fill_box = (183, 142, 521, 848)  
x, y, w, h = dog_bbox
dog_region = img[y:y+h, x:x+w]
fill_x, fill_y, fill_w, fill_h = dog_fill_box
fill_region = img[fill_y:fill_y+fill_h, fill_x:fill_x+fill_w]
modified_img = np.copy(img)
modified_img[y:y+h, x:x+w] = 0
if fill_w != w or fill_h != h:
    fill_region = cv2.resize(fill_region, (w, h))
modified_img[y:y+h, x:x+w] = fill_region
new_x = x + 600 
new_y = y  
if new_x + w > img.shape[1] or new_y + h > img.shape[0]:
    raise ValueError("New position is out of bounds.")
modified_img[new_y:new_y+h, new_x:new_x+w] = dog_region
cv2.imwrite("modified_dog_img.jpg", modified_img)
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
```

```python
blurred_image = cv2.GaussianBlur(img, (5, 5), 0)
plt.imshow(blurred_image, cmap="gray")
plt.title("Blurred Image")
plt.axis("off")
```
